```python
path = r'C:\Users\Janvi\Documents\09-2024 Instacart Basket Analysis'
```

```python
ords_prods_merge = pd.read_pickle(os.path.join(path, '02 Data', 'Prepared Data', 'merge_loop.pkl'))
```

```python
#calculating Mean for order number
```

```python
ords_prods_merge.groupby('department_id').agg({'order_number': ['mean']})
```

```python
##generating maximum orders for each users
```

```python
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
##looking at the max orderrs
```

```python
ords_prods_merge.head(15)
```

```python
##adding flag for loyal customers
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
ords_prods_merge['loyalty_flag'].value_counts(dropna= False)
```

```python
##looking at the data 
```

```python
ords_prods_merge.head()
```

```python
##finding spending habits of customers
```

```python
ords_prods_merge.groupby('loyalty_flag').agg({'prices': ['mean', 'min', 'max']})
```

```python
#defining high or low spender customers
```

```python
ords_prods_merge['mean_product_price'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
ords_prods_merge[['user_id', 'mean_product_price']].head(10)
```

```python
#marking as high or low spender
```

```python
ords_prods_merge.loc[ords_prods_merge['mean_product_price'] >= 10, 'spending_flag'] = 'High spender'
```

```python
ords_prods_merge.loc[ords_prods_merge['mean_product_price'] <= 10, 'spending_flag'] = 'Low spender'
```

```python
ords_prods_merge.head()
```

```python
##Finding frequency of purchasing for customer
```

```python
ords_prods_merge['median_days_between_orders'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
ords_prods_merge.head()
```

```python
ords_prods_merge.loc[ords_prods_merge['median_days_between_orders'] > 20, 'order_frequencey'] = 'Non-frequent customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['median_days_between_orders'] <= 20) & (ords_prods_merge['median_days_between_orders'] > 10), 'order_frequency'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['median_days_between_orders'] <= 10, 'order_frequencey'] = 'Frequent customer'
```

```python
ords_prods_merge.head()
```

```python
ords_prods_merge.drop(columns = ['order_frequency'])
```
